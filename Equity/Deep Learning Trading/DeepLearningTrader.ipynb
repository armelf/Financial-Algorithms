```python
import os
import pandas as pd
import numpy as np
import DataPPRL as DRL
import train as TR
```

```python
##############################################################################
# tensorflow  1.14.0
# Windows 10
##############################################################################

FSize           = 5     
PSize           = 2
PStride         = 2
NumAction       = 3



# hyper parameters described in the paper 
#################################################################################
maxiter         = 50000         # maxmimum iteration number         
learning_rate   = 0.00001       # learning rate
epsilon_min     = 0.1           # minimum epsilon

W               = 16           # input matrix size
M               = 300          # memory buffer capacity
B               = 10            # parameter theta  update interval               
C               = 300          # parameter theta* update interval ( TargetQ )
Gamma           = 0.99          # discount factor
P               = 0.35          # transaction penalty while training.  0.05 (%) for training, 0 for testing
Beta            = 32            # batch size
#################################################################################
```

```python
import tensorflow as tf

initial_path_train = 'YourSP500DataPath'+'/US/Train/'
initial_path_test = 'YourSP500DataPath'+'/US/Test/'
concatdates_train = '1999_2016'
concatdates_test = '2016_2020'

for companies_bundle in true_companies_bundles:
    print('')
    print('Companies bundle {} started'.format(companies_bundle))
    print('')
    
    # initialize
    DRead           = DRL.DataReaderRL()
    Model           = TR.trainModel(1.0, epsilon_min, maxiter, Beta, B , C, learning_rate, P)
    
    # Companies string
    string = ''
    for company in companies_bundle:
        string+='_'+str(company)
        
    ########## Train Data Import ############

    # folder path for training
    filepathX       =   initial_path_train+'W_'+str(concatdates_train)+'/inputX'+string+'.txt'
    filepathY       =   initial_path_train+'W_'+str(concatdates_train)+'/inputY'+string+'.txt'

    XData           = DRead.readRaw_generate_X(filepathX, W, W)                       # input chart
    YData           = DRead.readRaw_generate_Y(filepathY, len(XData), len(XData[0]))  # L_c^t  
    
    ########## Train Model ############
    tf.reset_default_graph()

    Model.set_Data(XData,YData)
    Model.trainModel(W,W, FSize, PSize, PStride, NumAction, M, Gamma)
    
    ########## Test Model ############
    tf.reset_default_graph()

    test_in = 'test'
    if test_in == 'train':
        bundle = concatdates_train
        print('')
        print(bundle)
        print('')
        folder = initial_path_train+'W_'+str(concatdates_train)+'/'
        filepathX       =   folder +'inputX'+string+'.txt'
        filepathY       =   folder +'inputY'+string+'.txt'
    else:
        bundle = concatdates_test
        print('')
        print(bundle)
        print('')
        folder = initial_path_test+'W_'+str(concatdates_test)+'/'
        filepathX       =   folder +'inputX'+string+'.txt'
        filepathY       =   folder +'inputY'+string+'.txt'
        
    sess,saver, state, isTrain, rho_eta = Model.TestModel_ConstructGraph(W,W,FSize,PSize,PStride,NumAction)

    XData           =   DRead.readRaw_generate_X(filepathX, W, W)
    YData           =   DRead.readRaw_generate_Y(filepathY, len(XData), len(XData[0]))   

    Model.set_Data(XData, YData)

    results_dir_path = folder + 'Results'
    if not os.path.exists(results_dir_path):
        os.makedirs(results_dir_path)   

    Model.Test_SeveralAssets_Prediction(sess, saver, state, isTrain, rho_eta, W, W, NumAction, results_dir_path, bundle, companies_bundle)
    
    print('')
    print('Companies bundle {} completed'.format(companies_bundle))
    print('')
```

```python
#Compute final PnL and equity curve
concatdates_test = '2016_2020'
method = 'SeveralAssetsPortfolio'
final_df = pd.DataFrame()
all_companies = list()
#P = 0.25 #50 bips round trip transaction costs
P = 0.0

def final_returns_df_computation(P):
    for companies_bundle in true_companies_bundles:
        string = ''
        for company in companies_bundle:
            string+='_'+str(company)

        results_string = 'Results_'+method+'_'+concatdates_test+string
        base_results_path_test = 'YourSP500DataPath'+'/US/Test/'
        results_path_test = base_results_path_test+'W_'+str(concatdates_test)+'/'+'Results/'
        result_df = pd.read_csv(results_path_test+results_string+'.csv')

        arrays = [list(result_df['Company']), list(result_df['Date'])]
        index = pd.MultiIndex.from_arrays(arrays, names=('Company', 'Date'))
        result_df.index = index

        for i, company in enumerate(companies_bundle):
            all_companies.append(company)
            result_df_company = result_df.loc[i]
            final_df['Prediction'+str(company)] = result_df_company['Predicted Signal']
            final_df['NextRets'+str(company)] = result_df_company['Tomorrow Returns']
            final_df['StratRets'+str(company)] = np.where(final_df['Prediction'+str(company)]==1,final_df['NextRets'+str(company)],0)

    #Percentages allocated everyday and final returns
    final_df['Strategy Returns'] = 0.
    for company in all_companies:
        final_df['Percentage'+company] = 0.

    indexes = final_df.index

    for d in range(len(final_df)):
        ret = 0
        count = 0
        TR = 0
        for company in all_companies:
            if final_df['Prediction'+company][d] == 1:
                ret+=final_df['StratRets'+company][d]
                count+=1
        for company in all_companies:
            if final_df['Prediction'+company][d] == 1:
                final_df['Percentage'+company][d] = 1/count
            else:
                final_df['Percentage'+company][d] = 0

        if count!=0:
            ret/=count

        if float(ret)!=float(0):
            final_df.loc[indexes[d],'Strategy Returns'] = ret

        print('Return day without TR {} is {}'.format(d, ret))


    final_df['Transactions'] = 0.
    totalTR = 0
    print('')
    for d in range(len(final_df)):
        TR = 0
        for company in all_companies:
            if d == 0:
                TR += abs((final_df['Percentage'+company][d]-0))
            else:
                TR += abs((final_df['Percentage'+company][d]-final_df['Percentage'+company][d-1]))
        final_df.loc[indexes[d],'Transactions'] = TR
        totalTR+=TR

        print('Total transactions of day {} is {}'.format(d, TR))

    print('')
    print('Total number of transactions is ', totalTR/2)
    print('')

    for d in range(len(final_df)):
        final_df.loc[indexes[d],'Strategy Returns']-=P*final_df['Transactions'][d]
        print('Return day with TR {} is {}'.format(d, final_df['Strategy Returns'][d]))
    
    return final_df
```

```python
#Final returns dataframe without transaction costs
P = 0.0 #0 pips round trip transaction costs
final_df = final_returns_df_computation(P)
```

```python
import matplotlib.pyplot as plt

#Cumulative returns
final_df['Cumulative Strategy Returns'] = np.cumsum(final_df['Strategy Returns'])

plt.figure(figsize=(10,5))
plt.plot(list(final_df['Cumulative Strategy Returns']), color='g', label='Strategy Returns')
plt.xlabel('Number of months')
plt.ylabel('Returns in %')
plt.legend()
plt.show()

#Sharpe Ratio
sharpe = np.sqrt(12)*np.mean(final_df['Strategy Returns'])/np.std(final_df['Strategy Returns'])
print('Sharpe Ratio is ', sharpe)
```
